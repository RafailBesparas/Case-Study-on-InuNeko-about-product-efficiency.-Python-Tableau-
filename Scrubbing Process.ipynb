### Question 2: Drop Duplicate Records
Drop the duplicated records and assign the output DataFrame to a new variable called `df_cleaned`. Use this from now on.

```python
df_cleaned = df.drop_duplicates()
df_cleaned.head()
```

### Question 3: Consistent Formatting

Fix the inconsistencies in the column `cust_state`.

```python
for x in df_cleaned.index:
    if df_cleaned.loc[x,"cust_state"] == 'NY':
        df_cleaned.loc[x,'cust_state'] = 'New York' 
    elif df_cleaned.loc[x, 'cust_state'] == 'NJ':
        df_cleaned.loc[x,'cust_state'] = 'New Jersey'
    elif df_cleaned.loc[x, 'cust_state'] == 'PA':
        df_cleaned.loc[x,'cust_state'] = 'Pennsylvania'
    elif df_cleaned.loc[x, 'cust_state'] == 'CT':
        df_cleaned.loc[x,'cust_state'] = 'Connecticut'
        
df_cleaned.head()
```

### Question 4: Fistful of Dollars I
Validate the prices in `prod_price`. Remove any rows that are obviously wrong.

```python
for x in df_cleaned.index:
    if df_cleaned.loc[x,"prod_price"] > 200:
        df_cleaned.drop(x, inplace = True)
        
df_cleaned.head()
```

### Question 5: Fistful of Dollars II
Correct the values in the `total_sales column`. Use prod_price and `trans_quantity`.

```python
for x in df_cleaned.index:
    df_cleaned.loc[x,"total_sales"] = df_cleaned.loc[x,'prod_price']*df_cleaned.loc[x,'trans_quantity']

df_cleaned.head()
```
